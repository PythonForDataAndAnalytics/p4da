```python
import pandas as pd
df = pd.read_csv('DQ-Example-Category.csv', index_col='id')
df
```

```python
df.children.unique()
```

```python
df['children_clean'] = df.children.str.lower()
df[['children', 'children_clean']]
```

```python
df['children_clean'] = df.children_clean.map({'y':1, 'n':0})
df[['children', 'children_clean']]
```

```python
df.gender.unique()
```

```python
df['gender_clean'] = df.gender.str.lower()
df['gender_clean'] = df.gender_clean.str[0]
df.gender_clean.astype('category')
df[['gender', 'gender_clean']]
```

```python
df['age_range'] = pd.cut(df.age, [0, 18, 21, 30, 100],
                         labels=['<18', '18-21','22-30','31+'])
df[['age','age_range']]
```

```python
regions = pd.read_csv('regions.csv')
regions
```

```python
df = df.merge(regions, on='state', how='left')
df[['state', 'region']]
```
