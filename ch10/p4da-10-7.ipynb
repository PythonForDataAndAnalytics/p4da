```python
dfAQI
```

```python
newrow = \
 pd.DataFrame({'concern':['Worst'],'low':[1000],'high':[9999],'number':7},
              index=['Brown'])
dfAQI = dfAQI.append(newrow)
```

```python
dfAQI
```

```python
dfAQI.drop('number', axis='columns', inplace=True)
dfAQI
```

```python
dfAQI.drop('Brown', inplace=True)
dfAQI
```
